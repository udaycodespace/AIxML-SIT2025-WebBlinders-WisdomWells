## üîç Objective

The goal of this project qualification assignment is to **perform various DataFrame manipulations and column-wise operations without using inbuilt shortcuts or high-level pandas utilities** like `.apply()` or `.dt`. The tasks focus on mathematical transformation, logical operations, and manual date extraction using basic Python and pandas features.


```python
## üöÄ Dataset Used
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(10, 4),
                  index=pd.date_range('2021-07-18', periods=10),
                  columns=['a', 'b', 'c', 'd'])
```

```python
df
```

üß™ Tasks Implemented

---

1Ô∏è‚É£ Create a new column and store the logarithmic values for column d

**Constraint:** Without using np.log() directly on Series

```python
def log_value(x):
    return x if x <= 0 else np.log(x)

df['log_d'] = [log_value(x) for x in df['d']]
```

```python
#To Show newly created column!
df
```

2Ô∏è‚É£ Create a new column where:


*   Negative values of column d remain same
*   Positive values are replaced with their log values

```python
df['transformed_d'] = [x if x < 0 else np.log(x) for x in df['d']]
```

```python
df
```

3Ô∏è‚É£ Create a column from column a that returns:


*   'yes' if the value is even (after converting to integer)
*   'no' otherwise


```python
df['even_a'] = ['yes' if int(x) % 2 == 0 else 'no' for x in df['a']]
```

```python
df
```

4Ô∏è‚É£ Create a new column with index values

```python
df['index_col'] = [i for i in df.index]
```

```python
df
```

5Ô∏è‚É£ Split index (date) into three new columns: year, month, and day

```python
df['year'] = [str(i).split('-')[0] for i in df.index]
df['month'] = [str(i).split('-')[1] for i in df.index]
df['day'] = [str(i).split('-')[2].split(' ')[0] for i in df.index]
```

```python
df
```

6Ô∏è‚É£ Create a column that returns the weekday name of the index manually (without using .day_name())

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['weekday'] = [days[i.weekday()] for i in df.index]
```
